```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
#sepal length와 sepal width만 선택
X = iris.data[:, [0, 1]]
X
```

```python
#K-means 클러스터링 적용
kmeans = KMeans(n_clusters = 3)
kmeans
```

```python
kmeans.fit(X)
```

```python
#각 데이터 포인트의 클러스터 레이블 얻기
labels = kmeans.labels_
labels
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels)
center = kmeans.cluster_centers_
# print(center)
plt.scatter(center[:, 0], center[:, 1], c="red", marker='X')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sa = silhouette_score(X[:, 0:1], labels)
print(sa) # -1 ~ 1
```

# 왼쪽 분류, 오른쪽 군집화 시각화

```python
data = load_iris()
X = data.data
y = data.target
data
```

```python
X
```

```python
plt.figure(figsize = (10, 6))

#왼쪽 : 분류
plt.subplot(121)
plt.plot(X[y==0, 2], X[y==0, 3], "yo", label="setosa")
plt.plot(X[y==1, 2], X[y==1, 3], "bs", label="versicolor")
plt.plot(X[y==2, 2], X[y==2, 3], 'g^', label="virginica")

#X[y==0, 2]
# 클래스 레이블이 0인 경우 해당하는 데이터 포인트 중에서 
# 2번째열에 해당하는 값을 가져옴

#오른쪽 : 군집화
plt.subplot(122)
plt.scatter(X[:, 2], X[:, 3], c='k')

plt.tick_params(labelleft=False)

plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=52)
kmeans.fit(X[:, 2:4])

#각 데이터 포인트에 대한 클러스터 레이블
labels = kmeans.labels_
labels

#k-means 중심점
center = kmeans.cluster_centers_
print(center)

#시각화
plt.figure(figsize = (10, 6))

#왼쪽 : 분류
plt.subplot(121)
plt.plot(X[y==0, 2], X[y==0, 3], "yo", label="setosa")
plt.plot(X[y==1, 2], X[y==1, 3], "bs", label="versicolor")
plt.plot(X[y==2, 2], X[y==2, 3], 'g^', label="virginica")

#오른쪽 : kmeans 적용한 군집화
plt.subplot(122)
plt.scatter(X[:, 2], X[:, 3], c = labels, marker='.')
plt.scatter(center[:, 0], center[:, 1], c='red', marker='X', s=50)

plt.show()

```
