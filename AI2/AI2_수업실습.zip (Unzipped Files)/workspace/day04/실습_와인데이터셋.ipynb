```python
#와인 데이터셋 데이터 분석, 전처리과정_시각화, 모델 생성/학습/예측, 정확도확인
# 교차검증(3가지중 2가지) 정확도 확인

from sklearn.datasets import load_wine
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold

wine = load_wine()
wine
wine.keys()
wine.feature_names
#알콜
#말산
#회분
#회분의 알칼리도
#마그네슘
#총 폴리페놀
#플라보노이드 폴리페놀
#비 플라보노이드 폴리페놀
#프로안토시아닌
#색상의 강도
#색상
#희석 와인 od280/od315 비율
#프롤린
```

```python
print(wine.DESCR)

wine_df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
wine_df
```

```python
wine_df['target'] = wine.target
wine_df


wine_df.loc[wine_df.duplicated()]

# wine_df = wine_df.drop_duplicates()

wine_df.info()
wine_df.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.3, random_state=0)
X_train.shape, X_test.shape

```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt
```

```python
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc_rate = accuracy_score(y_test, y_pred)

print(acc_rate)
```
