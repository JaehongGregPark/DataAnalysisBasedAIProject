```python
import pandas as pd
from mlxtend import frequent_patterns
from mlxtend import preprocessing
```

```python
df = pd.read_csv('./chipotle.csv')
df
```

```python
#연관분석 함수를 사용하기 위해 리스트로 변환
df_temp = df[['order_id', 'item_name']]
# print(df_temp)

df_temp_arr = [[] for i in range(1835)]
# print(df_temp_arr)
num = 0
for i in df_temp['item_name']:
    df_temp_arr[df_temp['order_id'][num]].append(i)
    num += 1
    
# print(df_temp_arr)

#order_id는 1부터 시작하므로 빈값 제거
df_temp_arr.pop(0)
df_temp_arr
```

```python
df
```

```python
#한테이블에서 중복된 메뉴가 존재할 수 있으므로 set() 중복값을 제거후 list 반환
num = 0
for i in df_temp_arr:
    df_temp_arr[num] = list(set(df_temp_arr[num]))
    num += 1
df_temp_arr
```

```python
#preprocessing.TransactionEncoder 객체를 생성
#아이템 기반의 이진행렬로 변환
te = preprocessing.TransactionEncoder()
te

#fit() 메소드를 TrasactionEncoder를 데이터 프레임으로 학습시킨 후
#transform() 메소드를 통해 제 아이템 기반의 이진 행렬로 변환
te_arr = te.fit(df_temp_arr).transform(df_temp_arr)
te.columns_

#변환된 이진 행렬을 데이터프레임으로 변환후 각 열의 이름을 아이템이름으로 설정
# df1 = pd.DataFrame(te_arr, columns = te.columns_)
# df1
```
