```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings(action='ignore')
```

```python
df = pd.read_csv('/train.csv')
df
```

```python
#feature와 label 값 분리
features = df.drop('Survived', axis=1)
labels = df['Survived']
features
labels
```

```python
#features 데이터 전처리
#결측치제거, 불필요한 속성 제거, 레이블인코딩
print(features.isnull().sum())
```

```python
#결측치 제거(대체)
features['Age'].fillna(features['Age'].mean(), inplace=True)
features['Cabin'].fillna('N', inplace=True)
features['Embarked'].fillna('N', inplace=True)
```

```python
features.isnull().sum()
```

```python
features.head()
```

```python
#불필요한 속성제거(PassengerId, Name, Ticket)
features = features.drop(['PassengerId', 'Name', 'Ticket'], axis=1)
features
```

```python
#레이블 인코딩 진행
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
features['Sex'] = le.fit_transform(features['Sex'])
print(features['Sex'])
print(le.classes_)
features['Embarked'] = le.fit_transform(features['Embarked'])
print(le.classes_)
features['Cabin'] = le.fit_transform(features['Cabin'])
print(le.classes_)
features.head()
features.info()
```

```python
#train_test_split() 으로 features와 labels를 훈련세트와 테스트세트로 분할
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels,
                                                   test_size=0.2, random_state = 53)
#shape 확인
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#모델 학습 및 지표확인
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
#모델 생성
dt = DecisionTreeClassifier(random_state = 11)
dt
```

```python
#모델 학습
dt.fit(X_train, y_train)
```

```python
#예측
y_pred = dt.predict(X_test)
y_pred
```

```python
print(f"정확도 : {accuracy_score(y_test, y_pred)}")
```

```python
#교차검증
from sklearn.model_selection import cross_val_score
scores = cross_val_score(dt, features, labels, cv = 10)
```

```python
#교차 검증마다의 정확도 확인
for iter_count, accuracy in enumerate(scores):
    print(f"{iter_count + 1}번째 교차검증 정확도 : {accuracy}")

#평균정확도
print(f"평균 정확도 : {np.mean(scores)}")
```

# GridSearchCV 적용

```python
from sklearn.model_selection import GridSearchCV
```

```python
#의사결정나무 생성
tree_model = DecisionTreeClassifier()

#튜닝할 하이퍼파라미터 그리드 설정
param_grid = {
    'criterion' : ['gini','entropy'], #불순도 계산에 사용할 기준, criterion : 분할기준
    'max_depth' : [None, 10, 20, 30], #트리의 최대 깊이
    'min_samples_split' : [2, 5, 10], #노드를 분할하기 위한 최소 샘플수
    'min_samples_leaf' : [1, 2, 4] #리프노드가 되기 위한 최소 샘플수
}
```

```python
#gridSearchCV 객체 생성
grid_search = GridSearchCV(tree_model, param_grid, cv=5, scoring='accuracy')
grid_search
```

```python
#모델 훈련 및 최적의 파라미터 찾기
grid_search.fit(X_train, y_train)
```

```python
#최적의 하이퍼파라미터와 성능 출력
print(f"최적의 하이퍼파라미터 : {grid_search.best_params_}")
print(f"최적의 성능 : {grid_search.best_score_}")
```

# 랜덤포레스트 모델로 GridSearchCV 확인

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#튜닝할 하이퍼 파라미터 그리드 설정
param_grid = {
    'criterion' : ['gini','entropy'], #불순도 계산에 사용할 기준, criterion : 분할기준
    'max_depth' : [None, 10, 20, 30], #트리의 최대 깊이
    'min_samples_split' : [2, 5, 10], #노드를 분할하기 위한 최소 샘플수
    'min_samples_leaf' : [1, 2, 4] #리프노드가 되기 위한 최소 샘플수
}

#랜덤포레스트 모델 생성
rf_model = RandomForestClassifier()
rf_model
```

```python
#GridSearchCV 객체 생성
grid_search = GridSearchCV(rf_model, param_grid, cv = 5, scoring="accuracy")
grid_search
```

```python
#모델 훈련 및 최적의 하이퍼파라미터
grid_search.fit(X_train, y_train)
```

```python
#최적의 하이퍼파라미터와 성능 출력
print(f"최적의 하이퍼파라미터 : {grid_search.best_params_}")
print(f"최적 정확도 : {grid_search.best_score_}")
```
