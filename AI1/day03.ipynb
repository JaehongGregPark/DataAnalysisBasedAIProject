>bike 클래스 정의


```python
class Bike:
    def info(self, brand, model,color):
        self.brand = brand
        self.model = model
        self.color = color
        self.speed = 0

    def start(self):
        print(f"{self.model}의 자전가가 출발합니다")

    def stop(self):
        print(f"{self.color} 자전가가 정지합니다")
        self.speed = 0

    def increase_speed(self, amount):
        self.speed += amount
        print(f"{self.color} 자전거의 속도가 {self.speed}으로 변경되었습니다")


```

```python
bike1=Bike()

bike1.info("코리아","Mountain Bike","Blue")

print(bike1.color)
bike1.start()
bike1.stop()
bike1.increase_speed(10)
```

```python
bike2 = Bike()

bike2.info("korea", "Bike", "Red")
bike2.start()
```

##+) 매직매소드


```python
class MyIterator:
    def __init__(self,start,end):
        self.current = start
        self.end = end

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.end:
            print("값을 넣을수 없습니다")
            raise StopIteration #파이썬 2에서 사용
        else:
            result = self.current
            self.current += 1
            return result
```

```python
m = MyIterator(1,5)
print(m.current)

for num in m:
    print(num)
```

```python
class Student :
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"이름:{self.name}, 나이 : {self.age}"

    def __repr__(self):
        return f"Student({self.age})"
```

```python
st2 = Student("박재홍", 20)
print(st2)
```

```python
class MyClass:
    def __init__(self, value):
        self.value = value

    def __str__(self): #사용자에게 보여주는 목적으로 사용
        return f"MyClass의 값 : {self.value}"

    def __repr__(self): #디버깅 및 객체의 정확한 표현을 제공하기 위해 사용
        return f"MyClass({self.value})"
```

```python
obj = MyClass(2024)

print(obj)
print(str(obj))
print(repr(obj))
```

```python
#축구선수1명, 야구선수1명의 이름, 나이, 직업을 넣은 클래스 만들기

class SomeClass:
    #초기화 메소드(생성자)
    def __init__(self, name, age, job):
        self.name = name
        self.age = age
        self.job = job

    #introduce(self)
    def introduce(self):
        print(f"이름 : {self.name}\n나이 : {self.age}\n직업 : {self.job}")

    def __str__(self):
        return f"이름 : {self.name}\n나이 : {self.age}\n직업 : {self.job}"


user1 = SomeClass("류현진", 38, "야구선수")
user2 = SomeClass("손흥민", 33, "축구선수")
#메소드 호출까지진행
user1.introduce()
user2.introduce()

print(user1)
print(user2)

user2.country = "영국"
```

##2. 클래스변수와 인스턴스 변수

```python
#한국사람 클래스로 구성 => 이름, 나이, 주소는 인스턴스 변수
#한국 나라는 클래스 변수

class Korean:
        country = "한국"    #클래스변수
        def __init__(self, name, age, address) :
            self.name = name        #인스턴스변수
            self.age = age          #인스턴스변수
            self.address = address  #인스턴스변수

```

```python
man1 = Korean("홍길동",35,"서울")
print(man1.name)
print(man1.age)
print(man1.address)
print(man1.country) #객체명.클래스변수명 가능
print(Korean.country) #클래스명.클래스변수명 가능
# print(Korean.name) #인스턴스변수는 반드시 객체명.인스턴스변수명으로 사용해야한다
```

### 2)클래스 메소드


```python
class MyClass2:
    class_variable = 10 #클래스변수

    def __init__(self, value):
        self.value = value

    def instance_method(self):
        return f"인스턴스 메소드 : {self.value}"

    @classmethod
    def class_method(cls):
        print(cls)
        return f"클래스메소드 : {cls.class_variable}"

result = MyClass2.class_method()
print(result)

my2 = MyClass2(10)
print(my2.instance_method())


# <class '__main__.MyClass2'>
#__main__ : 클래스가 정의된 모듈을 나타낸다
#현재스크립트 또는 대화형세션에서 정의된클래스임을 의미
#MyClass2 : 클래스 이름
```

```python
#__main__ : 현재 스크립트가 실행중인 메인 프로그램을 나타낸다
#파이썬 스크립트 파일을 실행하면 해당 스크립트는 __main__ 모듈로 간주된다
#스크립트 파일이 다른 모듈에서 임포트 되는 것과 구분하기 위한 목적으로 사용된다

def hello():
    print("hello, world")

if __name__ == "__main__":
    hello()
```

## 4. 상속

```python
#슈퍼클래스
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass
```

```python
#서브클래스
class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name)
        self.color = color

    def speak(self):
        return f"{self.name}이 야옹!하고 웁니다"

    def eat(self, food):
        print(f"{self.name}이 {food}를 먹습니다")
```

```python
cat = Cat("별이", '노란색')
cat.eat("추르")
print(cat.speak())
```
