```python
# 2. 3명의 학생의 python 점수를 입력받고 3명의 평균 출력하기
#1) 총합을 저장할 벼수
#2) 반복문(for range(3))
#3) 점수입력받을 변수 = input()
#4) 총합 += 입력받은점수
#5) 평균계산(총합/3)
#6) 출력
#num1=int(input("점수를 입력하세요"))
#num2=int(input("점수를 입력하세요"))
#num3=int(input("점수를 입력하세요"))

#num = (num1 + num2 + num3) / 3
#print(num)

# 학생 수
num_students = 3

# 각 학생의 파이썬 점수를 저장할 리스트
python_scores = []

# 학생들의 파이썬 점수 입력 받기
for i in range(num_students):
    while True:
        try:
            score = float(input(f"학생 {i + 1}의 파이썬 점수를 입력하세요: "))
            # 점수가 0에서 100 사이인지 확인
            if 0 <= score <= 100:
                python_scores.append(score)
                break
            else:
                print("점수는 0에서 100 사이어야 합니다. 다시 입력해주세요.")
        except ValueError:
            print("올바른 숫자를 입력하세요.")

# 평균 계산
average_score = sum(python_scores) / num_students

# 평균 출력
print(f"\n3명의 학생들의 파이썬 점수 평균은 {average_score:.2f}입니다.")
```

```python
# 2. 3명의 학생의 python 점수를 입력받고 3명의 평균 출력하기

#1) 총합을 저장할 변수
#2) 반복문(for range(3))
#3)     점수입력받을 변수 = input()
#4)     총합 += 점수입력
#5) 평균 계산(총합/3)
#6) 출력

total = 0
for i in range(3):
    # print(i)
    score = int(input(f"{i+1}번 학생의 python점수 입력 : "))
    total += score #total = total + score

avg = total / 3
print(f"3명의 학생의 평균 python점수 : {avg}")
```
