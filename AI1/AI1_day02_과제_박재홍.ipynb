# Q4. 주어진 리스트의 합, 최댓값, 최솟값을 출력하기

```python
numbers_list = [10, 2, 8, 5, 15]

# 리스트의 합계 계산 및 출력
sum_value = 0
for num in numbers_list:
    sum_value += num
print(f"Sum: {sum_value}")

# 리스트의 최댓값 계산 및 출력
max_value = numbers_list[0]
for num in numbers_list[1:]:
    if num > max_value:
        max_value = num
print(f"Maximum: {max_value}")

# 리스트의 최솟값 계산 및 출력
min_value = numbers_list[0]
for num in numbers_list[1:]:
    if num < min_value:
        min_value = num
print(f"Minimum: {min_value}")

numbers_list = [10, 2, 8, 5, 15]

# 리스트의 합계 계산 및 출력
sum_value = sum(numbers_list)
print(f"Sum: {sum_value}")

# 리스트의 최댓값 계산 및 출력
max_value = max(numbers_list)
print(f"Maximum: {max_value}")

# 리스트의 최솟값 계산 및 출력
min_value = min(numbers_list)
print(f"Minimum: {min_value}")
```

# Q6. 주어진 문자열에서 특정 문자가 몇 번 등장하는지 횟수 출력하기

```python
text3 = "python programming is powerful and pythonic"
target_char = "p"

# 문자열에서 특정 문자가 나타나는 횟수 계산
count = text3.count(target_char)

# 결과 출력
print(f"주어진 문자열에서 문자열'{target_char}'는 {count} 번 등장합니다.")

# 문자열에서 특정 문자가 나타나는 횟수 계산
count = sum(1 for char in text3 if char == target_char)
# 결과 출력
print(f"주어진 문자열에서 문자열'{target_char}'는 {count} 번 등장합니다.")
```

# Q11. 사용자가 입력한 정수값을 리스트에 저장하고 최대값과 최소값, 합계를 반환하는 함수

# 매개변수 o, 리턴값 o

```python
def number_mix():
    # 리스트 컴프리헨션을 사용하여 사용자로부터 세 개의 정수 입력 받기
    numbers = [int(input("첫 번째 숫자를 입력하세요: ")), int(input("두 번째 숫자를 입력하세요: ")), int(input("세 번째 숫자를 입력하세요: "))]
    return max(numbers), min(numbers), sum(numbers)

result = number_mix()
print("최댓값:", result[0])
print("최솟값:", result[1])
print("합계:", result[2])


def number_mix2():
    # For 루프를 사용하여 사용자로부터 세 개의 정수 입력 받기
    numbers = []
    for i in range(3):
        numbers.append(int(input(f"{i + 1}번째 숫자를 입력하세요: ")))
    return max(numbers), min(numbers), sum(numbers)

result = number_mix()
print("최댓값:", result[0])
print("최솟값:", result[1])
print("합계:", result[2])

def number_mix3():
    # 리스트 컴프리헨션과 루프를 사용하여 사용자로부터 세 개의 정수 입력 받기
    numbers = [int(input(f"{i + 1}번째 숫자를 입력하세요: ")) for i in range(3)]
    return max(numbers), min(numbers), sum(numbers)

result = number_mix()
print("최댓값:", result[0])
print("최솟값:", result[1])
print("합계:", result[2])


def number_mix4():
    # While 루프를 사용하여 사용자로부터 세 개의 정수 입력 받기
    numbers = []
    i = 0
    while i < 3:
        numbers.append(int(input(f"{i + 1}번째 숫자를 입력하세요: ")))
        i += 1
    return max(numbers), min(numbers), sum(numbers)

result = number_mix()
print("최댓값:", result[0])
print("최솟값:", result[1])
print("합계:", result[2])

def number_mix5():
    # map() 및 split()을 사용하여 사용자로부터 세 개의 정수 입력 받기
    numbers = list(map(int, input("띄어쓰기로 구분된 세 숫자를 입력하세요: ").split()))
    return max(numbers), min(numbers), sum(numbers)

result = number_mix()
print("최댓값:", result[0])
print("최솟값:", result[1])
print("합계:", result[2])


```

# Q12. 리스트 평균 계산
# 함수명 : list_avg
# 매개변수 o, 반환값 o
- 매개변수 : 숫자로 이루어진 리스트
- 반환값: 리스트의 평균을 반환

```python
def list_avg(numbers):
    # 리스트가 비어 있는지 확인하여 0으로 나누는 오류를 방지합니다.
    if len(numbers) == 0:
        return None  # 특정 값을 반환하거나 예외를 발생시키도록 필요에 따라 수정할 수 있습니다.

    # 평균을 계산합니다.
    average = sum(numbers) / len(numbers)

    return average

my_list = [1, 2, 3, 4, 5]
result = list_avg(my_list)
print(f"리스트 {my_list}의 평균은: {result}")
```
