```python
#       A   B   C
# 0       1   6   11
# 1       2   7   12
# 2       3   8   13
# 3       4   9   14
# 4       5   10   15

df = pd.DataFrame({'A':[1,2,3,4,5,], 'B':[6,7,8,9,10], 'C':[11,12,13,14,15]})

#1. iloc를 사용
df
print(df.iloc[1:4])
#2. loc를 사용

print(df.loc[1:3])

# [1번과 2번의 결과]
#    A  B   C
# 1  2  7  12
# 2  3  8  13
# 3  4  9  14

#3. loc 사용, B열이 8보다 큰 모든 행을 출력
#    A   B   C
# 3  4   9  14
# 4  5  10  15

print(df.loc[df['B'] > 8])
#4. 다음 데이터를 포함하는 데이터 프레임을 생성하고 출력
# name      age       city
# 짱구       25        New York
# 철수       32        Paris
# 훈이       18        London
# 맹구       47        Seoul
# 유리       23        Seoul

data = {'name': ['짱구', '철수', '훈이', '맹구', '유리'],
        'age': [25, 32, 18, 47, 23],
        'city': ['New York', 'Paris', 'London', 'Seoul', 'Seoul']}

# 데이터프레임 생성
df = pd.DataFrame(data)

df


#5. loc 사용
print(df.loc[[1,2,3],['name','age','city']])
#[출력결과]
#       name  age    city
# 1     철수   32   Paris
# 2     훈이   18  London
# 3     맹구   47   Seoul


#6. iloc 사용
print(df.iloc[1:3,0:2])
#[출력결과]
#       name  age
# 1     철수   32
# 2     훈이   18
```

```python
df

#나이가 20이상 거주지 서울인 행만 추출
print(df[df['age']>= 20])
print(df[df['city'] == 'Seoul'])

print(df.loc[(df['age'] >= 20) & (df['city'] == 'Seoul')])
```
