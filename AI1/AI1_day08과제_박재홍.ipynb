```python
data = {
    '이름' : ['짱구', '철수', '유리', '맹구', "짱아", "짱아"],
    '성별' : ['남', '남', '여', '남', '여', '여'],
    '나이' : [15, 14, None, 10, 5, 5],
    '키' : [150.1, 163.2, 100.2, None, 80.3, 80.3],
    '몸무게' : [40, 43, 34, 38, None, None],
    '학생여부' : [True, True, False, None, False, False]
}

#1) 데이터 프레임 정보 확인
# 데이터프레임 생성
df = pd.DataFrame(data)
print("1) 데이터프레임 정보 확인:")
df.info()
#2) 성별열을 기준으로 그룹화 하여 키의 평균값 구하기
avg_h = df.groupby('성별')['키'].mean().round(2)
print("\n 2) 성별열을 기준으로 그룹화 하여 키의 평균값:\n", avg_h)

#3) 나이열을 기준으로 내림차순 정렬하기
#참고 : .sort_values('컬럼명') 오름차순 정렬
#       .sort_values('컬럼명', ascending = False) 내림차순 정렬
srt_a = df.sort_values('나이', ascending=False)
print("\n 3) 나이 열을 기준으로 내림차순 정렬:\n", srt_a)

#4) 키열과 몸무게열의 평균값 구하기(apply)
avg_hw = df[['키', '몸무게']].apply(lambda x: x.mean())
print("\n4) 키열과 몸무게열의 평균값 구하기(apply): \n", avg_hw)

#5) 성별열의 값을 '남'은 1로, '여'는 2로 변환하고 정보 출력하기
df['성별'] = df['성별'].map({'남': 1, '여': 2})
print("\n 5)  성별열의 값을 '남'은 1로, '여'는 2로 변환하고 정보 출력하기:")
print(df)

#6) 나이의 None값을 나이열의 최소값으로 채우기
df['나이'].fillna(df['나이'].min(), inplace=True)
print("\n 6) 나이의 None값을 나이열의 최소값으로 채우기:")
print(df)
#7) 키와 몸무게의 None값을 각 열의 평균값으로 채우기
df[['키', '몸무게']] = df[['키', '몸무게']].apply(lambda x: x.fillna(x.mean()))
print("\n 7) 키와 몸무게의 None값을 각 열의 평균값으로 채우기:")
print(df)
#8) 학생여부의 None값을 나이가 5살 이상이면 True로 5살 미만이면 False로 채우기
#df['학생여부'] = df.apply(lambda row: True if row['학생여부'] == 'None' and row['나이'] >= 5  else row['학생여부'] , axis=1)
df['학생여부'] = np.where((df['학생여부'].isna()) & ( df['나이'] >= 5), True, df['학생여부'])
print("\n8) 학생여부의 None값을 나이가 5살 이상이면 True로 5살 미만이면 False로 채우기:")
print(df)

#

#8) 학생여부의 None값을 나이가 5살 이상이면 True로 5살 미만이면 False로 채우기
df['학생여부'].fillna(df['나이']>= 5, inplace=True)
df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#!sudo apt-get install -y fonts-nanum #나눔폰트 설치
#!sudo fc-cache -fv
#!rm ~/.cache/matplotlib -rf
#설치 후 런타임->세션 다시시작 클릭
#현 코드셀 다시 실행
plt.rc('font', family = 'NanumBarunGothic')
# plt.rc('font', size = 12)

# channal.csv 파일
# 1) 데이터프레임 생성(정보확인하기)
df = pd.read_csv('channal.csv')
print(df)
print(df.info())
# 2) 연도별 각 채널의 시청률 변화 시각화(선그래프)
df.set_index('Unnamed: 0').plot(kind='line', marker='o')
plt.title('연도별 각 채널의 시청률 변화')
plt.xlabel('Year')
plt.ylabel('시청률')
plt.legend(title='채널', bbox_to_anchor=(1, 1))
plt.show()
# 3) 연도별 각 채널의 시청률 평균을 계산하고 막대그래프로 시각화
# 각 연도별 각 채널의 평균 시청률을 계산
average_ratings = df.set_index('Unnamed: 0').mean()

# 평균 시청률을 막대 그래프로 시각화
average_ratings.plot(kind='bar', color='skyblue')
plt.title('연도별 각 채널의 평균 시청률')
plt.xlabel('채널')
plt.ylabel('평균 시청률')
plt.show()
# 4) 연도별 각 채널의 시청률 변화를 확인하고 가장 큰 변화를 보인 채널 출력하기
# 'Unnamed: 0' 열을 인덱스로 설정하여 계산을 용이하게 함
df.set_index('Unnamed: 0', inplace=True)

# 각 채널별 시청률 변화를 계산
ratings_change = df.diff()

# 가장 큰 변화를 보인 채널을 찾음
max_change_channel = ratings_change.abs().sum(axis=0).idxmax()

# 최대 변동을 보인 채널을 출력
print(f"최대 변동을 보인 채널: {max_change_channel}")

```
