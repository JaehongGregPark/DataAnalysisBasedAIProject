## 1. 넘파이

### 1) 넘파이 배열의 사칙연산


```python
import numpy as np
#2행2열
ar1 = np.array([[10, 10], [20, 20]])
ar2 = np.array([[10, 20], [30, 40]])

print(ar1, ar1.shape, ar1.ndim)
print(ar2, ar2.shape, ar2.ndim)

print(f"ar1+ar2 = \n{ar1 + ar2}")
print(f"ar1-ar2 = \n{ar1 - ar2}")
print(f"ar1*ar2 = \n{ar1 * ar2}")
print(f"ar1/ar2 = \n{ar1 / ar2}")
```

## 3. 시리즈

### 1) 기본구조

```python
import pandas as pd

s = pd.Series(['AI', 2024, 1.28],name="AI1 6일차", index=['과목명','연도','월일'])
print(s, type(s))
```

### 2) 리스트, 튜플 -> 시리즈 변환


```python
num = [1, 2, 3, 4, '5']
student = 'AI', '2024', 1.28, True
print(num, type(num))
print(student, type(student))

sr1 = pd.Series(num, name = "숫자", index = ['일', '이', '삼', '사', '오'])
print(sr1)

print(sr1.index, type(sr1.index))
print(sr1.values, type(sr1.values)) #시리즈의 값을 넘파이 배열로 가져옴
print(sr1.dtype, type(sr1.dtype))
```

```python
student = 'AI', '2024', 1.28, True
print(student, type(student))

sr2 = pd.Series(student, index = ['과정명', '년도','수강일', '강의수강여부'], name = '학생')
print(sr2)

#원소 선택(방법2)
#방법1 : 정수형 위치 인덱스를 사용하는 방법 => 리스트와 같은 원리
print(sr2[0], type(sr2[0]))
print(sr2[2], type(sr2[2]))
print(sr2[3], type(sr2[3]))

#방법2 : 인덱스 이름(라벨)을 사용하는 방법 => 딕셔너리와 같은 원리
print(sr2['강의수강여부'])

#여러값을 인덱싱
print(sr2[0:3],"\n") #슬라이싱을 이용한 방법(대괄호 한번만작성)
print(sr2[[0,3,2]], "\n") #띄어져있는 값을 추출할 때(대괄호 두번작성)
print(sr2[['강의수강여부']], type(sr2[['강의수강여부']]))
```

### 4) 시리즈 -> 딕셔너리 or 리스트 변환



```python
data = [1, 2, 3, 4]
sr = pd.Series(data, index = ['a', 'b', 'c', 'd'])
print(sr, type(sr))

#1) 파이썬의 함수를 이용하는 방법
#시리즈 -> 딕셔너리
di = dict(sr)
print(di, type(di)) #인덱스가 키가 된다

#시리즈 -> 리스트
li = list(sr)
print(li, type(li)) #시리즈의 인덱스값은 사용되지 않는다

```

```python
data = {'a': 10, 'b':20, 'c':30}
print(data, type(data))

sr = pd.Series(data)
print(sr, type(sr))

#판다스의 메소드로 변환하는 방법
#시리즈 -> 딕셔너리
di = sr.to_dict()
print(di, type(di))

#시리즈 -> 리스트
li = sr.tolist()
print(li, type(li))
```

### 5) 데이터프레임 -> 딕셔너리


```python
student = {'math':[90, 10], 'eng':[90, 10], 'kor':[100, 10], 'his':[80, 10]}
df = pd.DataFrame(student, index = ['학생1', '학생2'])

#.to_dict() : 컬럼은 키로, 레코드는 밸류, 인덱스는 키로
#{컬럼0:{인덱스0: 값, 인덱스1 :값}, 컬럼1:{인덱스0: 값, 인덱스1 :값}}

di1 = df.to_dict()
print(di1)

# .to_dict('list') : 컬럼은 키로, 각 쌍은 리스트로 변환
di2 = df.to_dict('list')
print(di2)

# .to_dict('index') : 인덱스는 키로, 레코드는 밸류, 컬럼은 키로
#{인덱스 : {컬럼0:값, 컬럼1:값, 컬럼2:값, 컬럼3:값} }
di3 = df.to_dict('index')
print(di3)

# .to_dict('records') : 각 행을 딕셔너리로 묶고, 리스트로 묶음
di4 = df.to_dict('records')
print(di4)

df
```
