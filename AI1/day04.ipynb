## 1. 상속


```python
#슈퍼클래스
class Animal:
    def __init__(self,name, age ) :
        self.name = name
        self.age = age

    def info(self) :
        print(f"{self.name}은 {self.age}살입니다")

    def eat(self, food):
        print(f"{self.name}이 {food}를 먹습니다")

    def sleep(self):
        print(f"{self.name}이/가 잠을 잡니다")
```

```python
#서브클래스
class Cat(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)
        self.breed = breed

    def speak(self):
        print(f"{self.name}이/가 야옹소리를 냅니다")
```

```python
#서브클래스
class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)
        self.breed = breed

    def speak(self):
        print(f"{self.name}이/가 멍멍소리를 짖습니다")
```

```python
#cat1 Animal클래스의 인스턴스
cat1 = Animal("별이", 5)
cat1.info()
cat1.sleep()
```

```python
cat2 = Cat("달이",6,"샴")
print(cat2)
cat2.info()
cat2.speak()
```

```python
dog1 = Dog("멍멍이",2,"골든리트리버")
dog1.info()
dog1.speak()
```

### 2) 서브클래스의 인스턴스 자료형



```python
class Animal:
    pass

class Dog(Animal):
    pass

class Cat(Animal):
    pass
```

```python
animal = Animal()
dog = Dog()
cat = Cat()

print(animal)
print(dog)
print(cat)

print(isinstance(animal, Animal))
#animal 객체는 Animal클래스의 타입이니? True, False
print(isinstance(animal, Cat))
#animal 객체는 슈퍼클래스의 객체, Cat 서브클래스
#animal 객체로는 Cat 서브클래스에 있는 속성를 사용할 수 없다

print(isinstance(cat, Cat)) #True
print(isinstance(cat, Animal)) #True
```

## 2. 예외처리


### 1) 예외처리의 필요성


```python
#1) 고전적인 예외처리
num1 = 3
num2 = 0
print(num1 / num2)
print("출력끝")
```

```python
#2) 모든 예외를 처리하는 방법
#if~else문으로 작성가능
num1 = input("숫자1입력 : ")
num2 = input("숫자2입력 : ")
if num2 == 0:
    print("0으로 나누는것은 불가능합니다")
else:
    print(f"{num1} / {num2} = {num1/num2}")

print("출력끝")
```

### 2) 예외처리 구조


```python
try:
    num1 = 10
    num2 = input("숫자입력 : ")
    print(num1 / num2)
except ZeroDivisionError as e:
    print(f"예외발생 : {e}")
    print("예외가 발생했습니다")
except TypeError as e:
    print(f"예외발생 : {e}")
    print("타입 에러가 발생했습니다")
print("예외처리 끝")
```

### 2) try ~ except ~ else ~ finally


```python
print("프로그램 시작")
    #   0    1    2    3    4
arr = ['h', 'e', 'l', 'l', 'o']
print(arr, type(arr))
print(arr[0]) #h
print(arr[9]) #IndexError: list index out of range

print("프로그램 끝")
```

```python
print("프로그램 시작")
try:
    arr = ['h', 'e', 'l', 'l', 'o']
    # print(arr[0])
    print(arr[9])
    print("try문 실행")
except TypeError as e:
    print(f"참조변수 출력 : {e}")
    print("except문 실행1")
except Exception as e:
    print(f"참조변수 출력 : {e}")
    print("except문 실행2")
else:
    print("else문 실행")
finally:
    print("finally문 실행")

print("프로그램 종료")
```

### (3) raise 예외클래스() 또는 예외클래스("문장")


```python
number = int(input("1부터 5까지 숫자 중 입력 : "))

# print(0 > -100)

if 0 > number or number > 5:
    # print("1부터 5까지 중의 숫자를 입력하세요")
    raise ValueError("1부터 5까지의 숫자만 입력하라고 했잖아요")
else:
   print(f"입력한 값은 {number}입니다")

print("출력 끝")
```

```python
#사용자 정의 예외 클래스 정의

class MyException(Exception):
    def __init(self, message):
        super().__init__(message)

my = MyException("예외")
print(my)
```

```python
def divide(a, b):
    if b == 0:
        raise MyException("0으로 나눌수없습니다")

    return a / b
```

```python
try:
    number = int(input("1부터 5까지 숫자 중 입력 : "))
    if 0 > number or number > 5:
        # print("1부터 5까지 중의 숫자를 입력하세요")
        raise ValueError("1부터 5까지의 숫자만 입력하라고 했잖아요")
    else:
        print(f"입력한 값은 {number}입니다")
except ValueError as e:
    print(f"예외메시지출력 : {e}")

print("출력 끝")
```

### 2) with 문을 이용한 파일생성


```python
with open('test03.txt', 'wt', encoding='utf-8') as f:
    f.write('김영선\n')
    f.write('AI프로젝트반1\n')
    f.write('오늘은 python 마지막시간입니다')
    print('test03.txt 파일이 생성되었습니다')


with open('test03.txt') as f:
    # data = f.read(10) #최대 10개의 문자를 읽어온다
    data = f.readlines()
    for line in data:
        print(line.strip()) #strip() 불필요한 문자열, 문자개행문자 제거된다

    print(data)
    print("test03.txt파일을 읽어왔습니다")
```
