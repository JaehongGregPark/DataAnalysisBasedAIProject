## 3.배열구조


### 1) list구조

```python
li1 = [1,2,3,'a','b',True]
print(li1, type(li1))
print(li1[0],type(li1[0]))
print(li1[0], type(li1[0]))
print(li1[-1], type(li1[-1]))
print(li1[-2], type(li1[-2]))

li2 = [[1, 2, 3], [1], 12]
print(li2, type(li2))
print(li2[0], type(li2[0]))
print(li2[1], type(li2[1]))
print(li2[1][0], type(li2[1][0]))
print(li2[0][2], type(li2[0][2]))

print(li1 + li2) #연결
# print(li1 - li2)
print(li1 * 5) #반복
```

## 6. 벡터



```python
import numpy as np

v = np.array([10])
print(v, type(v))

#.ndim 차원
print(v.ndim)

#.shape
print(v.shape) #열 (2, )

#2차원 벡터 => 배열
v1 = np.array([[1], [2]])
print(v1.ndim)
print(v1.shape) #행, 열
print(v1)
```

```python
#1차원 배열 전용 생성함수(arange(시작, 종료, 스텝값))

#range(시작, 종료, 스텝값) => 정수로만 만들수 있음

print(list(range(0, 10, 2)))
# print(list(range(0, 10, 0.5)))

v2 = np.arange(0, 10, 0.5)
print(v2, type(v2))
```

```python
v3 = np.arange(3,5)
print(v3)
print(v3.shape)#(행,열)(열,)

#노름 구하기
v_p = np.power(v3,2)
print(v_p, type(v_p))
v_r = np.sqrt(sum(v_p))
print(v_r)


#피타고라스 정리를 구하는 함수hypot()

print(np.hypot(v3[0],v3[1]))

#더 많은 원소를 가진 벡터를 가지고 크기를 구할 때는
#np.linalg 모듈 norm()를 이용함

print(np.linalg.norm(v3))
```

## 9. 넘파이 함수

```python
import numpy as np

#넘파이 배열 만들때 np.array([])
ar1 = np.array([1,2,3,4,5])
print(ar1, type(ar1))
ar2 = np.array([[1, 2, 3], [4, 5, 6]])
print(ar2, type(ar2))

#크기(형상)
print(f"벡터 : {ar1.shape}")
print(f"매트릭스 : {ar2.shape}")

#차원
print(f"벡터 : {ar1.ndim}")
print(f"매트릭스 : {ar2.ndim}")

#요소갯수
print(f"벡터 : {ar1.size}")
print(f"매트릭스 : {ar2.size}")

#데이터 타입
print(f"벡터 : {ar1.dtype}")
print(f"매트릭스 : {ar2.dtype}")

print(ar1)
print(ar2)
```

```python
#넘파이 추출(검색)
#벡터2라는 값만 추출
print(ar1[1])

#매트릭스에서 2라는 값만 추출
print(ar2[0][1])

#배열 슬라이싱
#매트릭스 첫번째 행 두번 째 열에 원소에 접근
print(ar2[0, 1])

```

```python
#수정
print(ar1)
print(ar2)

#벡터
ar1[1] = 20
print(ar1)

ar2[0][2] = 30
print(ar2)

ar2[1, 1] = 40.1
print(ar2)

# 수정시는 기존에 있는 자료형으로만 변경됨
# 원래 배열의 데이터 타입을 유지하기 위해 설계 되었기 때문에
# 데이터의 무결성과 데이터 타입의 안정성을 유지하기 위함

```

```python
#추가
print(ar2)
print(ar2.shape)

ar3 = np.array([[0, 0, 0], [0, 0, 0]])
print(ar3.dtype)
print(ar3.ndim)
print(ar3.shape)
print(ar3)
# arr = np.append(ar2, ar3, axis = 0)
#axis 축을 의미 , 0  수직축을 의미 => 행이 증가(열의 개수가 맞아야 함)
#                 1  수평축을 의미 => 열이 증가(행의 개수가 맞아야 함)
arr = np.append(ar2, ar3, axis = 1)
print(arr)


```
