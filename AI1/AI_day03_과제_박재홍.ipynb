# Q1. Student 클래스
- 데이터 속성
    
    - name: 학생 이름
    - age: 학생 나이
    - courses: 수강 중인 과목 목록 (리스트로 표현)

- 메소드
    
    add_course(course): 학생이 새로운 과목(course)을 수강하게 하고 courses 목록에 추가하는 메소드
    
    remove_course(course): 학생이 수강 중인 과목(course)을 삭제하고 courses 목록에서 제거하는 메소드

    list_courses(): 학생이 수강 중인 과목 목록을 출력하는 메소드

    get_age(): 학생의 나이를 반환하는 메소드를 작성

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.courses = []

    def add_course(self, course):
        if course not in self.courses:
            self.courses.append(course)
            print(f"{self.name}님이 {course} 수강 신청하였습니다.")

    def remove_course(self, course):
        if course in self.courses:
            self.courses.remove(course)
            print(f"{self.name}님이 {course} 수강 철회하였습니다.")
        else:
            print(f"{self.name}님은 현재 {course}을(를) 수강 중이 아닙니다.")

    def list_courses(self):
        print(f"{self.name}님의 수강 중인 과목 목록: {', '.join(self.courses)}")

    def get_age(self):
        return self.age


# 사용 예시
student1 = Student("홍길동", 20)
student1.add_course("수학")
student1.add_course("영어")
student1.list_courses()
student1.remove_course("과학")
student1.remove_course("수학")
student1.list_courses()

age_of_student1 = student1.get_age()
print(f"{student1.name}님의 나이는 {age_of_student1}세 입니다.")
```

#Q2. Student 클래스 확장하기
- 클래스 변수 school_name을 추가하고, 클래스 메소드 get_school_name(cls)을 작성하여 학교 이름을 반환
- 정적 메소드 is_adult(age)를 작성하여 학생의 성인 여부를 확인(성인은 18세 이상)

```python
class Student:
    school_name = "예시 학교"  # 클래스 변수

    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.courses = []

    def add_course(self, course):
        if course not in self.courses:
            self.courses.append(course)
            print(f"{self.name}님이 {course} 수강 신청하였습니다.")

    def remove_course(self, course):
        if course in self.courses:
            self.courses.remove(course)
            print(f"{self.name}님이 {course} 수강 철회하였습니다.")
        else:
            print(f"{self.name}님은 현재 {course}을(를) 수강 중이 아닙니다.")

    def list_courses(self):
        print(f"{self.name}님의 수강 중인 과목 목록: {', '.join(self.courses)}")

    def get_age(self):
        return self.age

    @classmethod
    def get_school_name(cls):
        return cls.school_name

    @staticmethod
    def is_adult(age):
        return age >= 18


# 사용 예시
student1 = Student("홍길동", 20)
student2 = Student("김철수", 16)

print(f"{student1.name}님의 학교 이름: {student1.get_school_name()}")
print(f"{student2.name}님의 학교 이름: {student2.get_school_name()}")

print(f"{student1.name}님은 성인입니까? {Student.is_adult(student1.age)}")
print(f"{student2.name}님은 성인입니까? {Student.is_adult(student2.age)}")
```

#Q3. 은행 계좌 클래스인 BankAccount를 작성
데이터 속성
- balance: 계좌 잔액 (처음에 0으로 초기화됨)

메소드
- deposit(amount): 계좌에 금액을 입금하고 잔액을 업데이트
- withdraw(amount): 계좌에서 금액을 출금하고 잔액을 업데이트합니다. 잔액보다 많은 금액을 출금하려고 하면 "잔액이 부족합니다." 메시지 반환
- get_balance(): 현재 계좌 잔액을 반환

```python
class BankAccount:
    def __init__(self):
        self.balance = 0

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"{amount}원이 입금되었습니다. 현재 잔액: {self.balance}원")
        else:
            print("입금 금액은 0보다 커야 합니다.")

    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                print(f"{amount}원이 출금되었습니다. 현재 잔액: {self.balance}원")
            else:
                print("잔액이 부족합니다.")
        else:
            print("출금 금액은 0보다 커야 합니다.")

    def get_balance(self):
        return self.balance


# 사용 예시
account = BankAccount()

account.deposit(1000)
account.withdraw(500)
account.withdraw(700)  # 잔액이 부족하여 출금 실패
print("현재 잔액:", account.get_balance())
```

```python
class BankAccount:
    def __init__(self):
        self.balance = 0

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"{amount}원을 입금했습니다")
        else:
            print("입금금액은 0보다 커야합니다")

    def withdraw(self, amount):
        if amount > 0:
            if self.balance >= amount:
                self.balance -= amount
                print(f"{amount}원을 출금했습니다")
            else:
                print("잔액이 부족합니다")
        else:
            print("출금금액은 0보다 커야합니다")

    def get_balance(self):
        return self.balance

#은행계좌 인스턴스 생성
account = BankAccount()

print(f"현재 계좌 잔액 : {account.get_balance()} ")

account.deposit(1000)
account.deposit(10000)
print(f"현재 계좌 잔액 : {account.balance}")

account.withdraw(5000)
print(f"현재 계좌 잔액 : {account.get_balance()}")
```
