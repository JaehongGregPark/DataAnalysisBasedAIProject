# 3차원 iris 그래프(t-SNE사용)

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
```

```python
iris = load_iris()
df = pd.DataFrame(data = np.c_[iris.data, iris.target])
df
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target']
df
```

```python
#3차원 t-SNE
tsne_df = TSNE(n_components = 3)
tsne_df = tsne_df.fit_transform(df)
tsne_df = pd.DataFrame(tsne_df,columns = ['component 0', 'component 1', 'component 2'])
tsne_df
```

## 3차원 그래프 시각화

```python
figure = plt.figure(figsize = (12, 8))
axes = figure.add_subplot(111, projection = "3d")

#target
tsne_df['target'] = df['target']
tsne_df

#target별 분리
tsne_0 = tsne_df[tsne_df['target'] == 0]
tsne_1 = tsne_df[tsne_df['target'] == 1]
tsne_2 = tsne_df[tsne_df['target'] == 2]

#target 별 시각화
axes.scatter(tsne_0['component 0'], tsne_0['component 1'], tsne_0['component 2'], label = '0')
axes.scatter(tsne_1['component 0'], tsne_1['component 1'], tsne_1['component 2'], label = '1')
axes.scatter(tsne_2['component 0'], tsne_2['component 1'], tsne_2['component 2'], label = '2')
plt.legend()
plt.show()
```
