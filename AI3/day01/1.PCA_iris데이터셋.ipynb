# 1. 데이터셋 불러오기

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
```

```python
#데이터셋 로드
iris = load_iris()
df = pd.DataFrame(data=iris.data)
#iris 데이터를 Numpy 배열에서 데이터 프레임으로 변환
df
```

## 데이터 셋 정규화

```python
#데이터셋 정규화
scaler = StandardScaler()
scaler.fit(df)
# 데이터 정규화를 위해 StandardScaler 모델을 입력 데이터에 맞춘다

df_scaled = scaler.transform(df)
#입력데이터를 정규화해서 변환
df_scaled
```

```python
# [-9.00681170e-01,  1.01900435e+00, -1.34022653e+00, -1.31544430e+00]
# 정규화된 데이터를 의미
# 각 행은 하나의 데이터 포인트를 의미, 각열은 해당 데이터 포인트의 특성을 의미
# 모든 특성이 평균이 0, 표준편차가 1로 표준화
```

```python
iris
```

```python
# 정규화된 값을 데이터 프레임으로 변환
df_scaled = pd.DataFrame(df_scaled)
df_scaled

#sepal_length, sepal_width, petal_length, petal_width컬럼명변경
df_scaled.columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
df_scaled['target']= iris.target
df_scaled
```

```python
df_scaled.describe() # 표준화된 데이터 프레임의 통계적 요약정보
```

```python
from sklearn.decomposition import PCA
```

```python
#2차원으로 차원축소
pca = PCA(n_components=2)
pca
```

```python
pca.fit(df_scaled.iloc[:,:-1])

pca_df = pca.transform(df_scaled.iloc[:,:-1])
pca_df = pd.DataFrame(pca_df, columns=['component 0', 'component 1'])
pca_df
```

```python
#PCA 설명력 출력
print(pca.explained_variance_ratio_)
```

## PCA 차원축소 결과 시각화

```python
import matplotlib.pyplot as plt

#target 불러오기
pca_df['target'] = df_scaled['target']
pca_df
```

```python
# target별 분리
pca_0 = pca_df[pca_df['target']==0]
pca_1 = pca_df[pca_df['target']==1]
pca_2 = pca_df[pca_df['target']==2]

# target별 시각화
plt.scatter(pca_0["component 0"], pca_0["component 1"], color="red", label="setosa")
plt.scatter(pca_1["component 0"], pca_0["component 1"], color="green", label="versicolor")
plt.scatter(pca_2["component 0"], pca_0["component 1"], color="blue", label="virginica")

plt.xlabel('component 0')
plt.ylabel('component 1')
plt.legend()
plt.show()
```
